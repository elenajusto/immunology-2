#### Channel Information

##### FSC-A (Forward Scatter Area):

- Purpose: Measure cell size.

- Macrophage Identification: Macrophages are typically larger than lymphocytes and other immune cells. Use FSC-A to isolate larger cell populations for further analysis.

##### FSC-W (Forward Scatter Width):

- Purpose: Differentiate between single cells and cell aggregates/clumps.

- Macrophage Identification: Helps ensure that you're analyzing single macrophages rather than aggregates by identifying their unique width profiles.

##### SSC-A (Side Scatter Area):

- Purpose: Measure granularity and internal complexity.

- Macrophage Identification: Macrophages tend to have higher SSC-A values due to their granular nature (e.g., vacuoles and lysosomes). Use SSC-A in combination with FSC-A to gate macrophages.

##### SSC-W (Side Scatter Width):

- Purpose: Differentiate cells based on their granularity width.

- Macrophage Identification: Helps distinguish between populations with similar complexity but different width profiles, ensuring you target the correct macrophage subpopulation.

##### PE-A (Phycoerythrin Area):

- Purpose: Detect fluorescence from PE-tagged antibodies.

- Macrophage Identification: Use antibodies against CD11b, F4/80 (mouse), or CD68 (human) conjugated to PE to identify macrophages. Positive PE-A signal indicates the presence of these macrophage-specific surface markers.

##### APC-A (Allophycocyanin Area):

- Purpose: Detect fluorescence from APC-tagged antibodies.

- Macrophage Activity: Use APC-labeled antibodies to detect activation markers like CD86 (M1 macrophage) or CD206 (M2 macrophage). Increased APC-A signals indicate macrophage activation or polarization state.

##### APC-Cy7-A (APC-Cy7 Tandem Dye Area):

- Purpose: Detect fluorescence from APC-Cy7-labeled antibodies.

- Macrophage Activity: Use this channel for additional markers of macrophage function or differentiation, such as MHC-II for antigen presentation or cytokines like TNF-Î± or IL-10 to assess inflammatory or regulatory functions. APC-Cy7-A can help in tracking multiparametric data without overlapping with other fluorochromes.

## Load Standards

```python
# Specify the directory containing your standard files
standard_directory = 'data_macrophage/CBA/fcs/standard'

# List of standard files (sorted for consistency)
standard_files = sorted([f for f in os.listdir(standard_directory) if f.endswith('.fcs')])

# Load each file and manually assign IDs
standard_measurements = {}
for i, standard_file in enumerate(standard_files, start=1):
    file_path = os.path.join(standard_directory, standard_file)
    sample_id = f'std {i}'  # Assigning IDs like 'std 1', 'std 2', etc.
    standard_measurements[sample_id] = FCMeasurement(ID=sample_id, datafile=file_path)

# Visualize the loaded standard measurements
print(standard_measurements)
```

### Visualise SSC-A vs FSC-A for Standards

```python
# Create a dictionary with file names as keys and concentrations as values
concentrations = {
    'std 1.fcs': 2500,
    'std 2.fcs': 1250,
    'std 3.fcs': 625,
    'std 4.fcs': 312,
    'std 5.fcs': 160,
    'std 6.fcs': 80,
    'std 7.fcs': 40,
    'std 8.fcs': 20,
    'std 9.fcs': 10,
    'std 10.fcs': 5,
    'std 11.fcs': 2.5,
    'std 12.fcs': 0
}

# Assuming you have your standard files in a dictionary like 'standard_measurements'
for standard in standard_measurements:
    # Strip the ".fcs" from the standard if necessary to match keys
    standard_key = standard if standard in concentrations else standard.replace('.fcs', '')
    
    # Get the corresponding concentration
    concentration = concentrations.get(standard_key, 1)  # Default to 1 if not found
    
    # Extract the relevant data (SSC-A and FSC-A), then multiply by the concentration
    ssc_a = standard_measurements[standard]['SSC-A'] * concentration
    fsc_a = standard_measurements[standard]['FSC-A'] * concentration
    
    # Plot the data
    plt.figure()  # Create a new figure for each standard
    plt.hist(ssc_a, bins=100, alpha=0.7, label='SSC-A')
    plt.hist(fsc_a, bins=100, alpha=0.7, label='FSC-A')
    plt.title(f'Standard: {standard} (Concentration: {concentration} pg/mL)')
    plt.legend()
    plt.show()  # Display the plot
```
